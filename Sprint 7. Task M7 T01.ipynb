Here is the explanation of all the columns in the Wine recognition dataset:

1. **Class**: Class or type of wine. It represents the category to which the wine belongs (1, 2, or 3). The class indicates the specific classification or grouping of the wine samples.
2. **Alcohol**: Represents the alcohol content of the wine samples, typically expressed as a percentage of alcohol by volume.
3. **Malic acid**: Indicates the amount of malic acid present in the wine.
4. **Ash**: Represents the ash content in the wine, which is the residue left after complete combustion.
5. **Alcalinity of ash**: Refers to the measure of the alkalinity or basicity of the ash in the wine.
6. **Magnesium**: Indicates the quantity of magnesium present in the wine.
7. **Total phenols**: Represents the total concentration of phenolic compounds in the wine, which contribute to its color and taste.
8. **Flavanoids**: Refers to the concentration of flavonoids, a type of phenolic compound, in the wine.
9. **Nonflavanoid phenols**: Represents the concentration of phenolic compounds other than flavonoids in the wine.
10. **Proanthocyanins**: Indicates the concentration of proanthocyanidins, a type of phenolic compound, in the wine.
11. **Color intensity**: Represents the color intensity or darkness of the wine.
12. **Hue**: Indicates the hue or shade of the wine color.
13. **OD280/OD315 of diluted wines**: Represents the optical density measurements of the wine at two different wavelengths, indicating its color density.
14. **Proline**: Indicates the concentration of proline, an amino acid, in the wine.

## Preprocessing:

```python
df[['Magnesium', 'Proline']] = df[['Magnesium', 'Proline']].astype('float')
df['class'] = df['class'].astype('category')
df.info()
```

```python
from scipy import stats
print('Null hypothesis H0, the sample has a normal distribution, It is Guessian')
print('Alternative hypothesis H1, the sample does not  have a normal distribution, It is not Guessian')

for i in df.columns:
    shapiro_test = stats.shapiro(df[i])
    print('For the variable ', i, 'with a p-value of ', shapiro_test.pvalue)
    if shapiro_test.pvalue>0.05:
        print('We cannot reject the null hypothesis, It is Guessian')
    else:
        print('We can reject the null hypothesis, It is not Guessian')
```

##### Among all the columns, just 'Alcalinity of ash' with a p-value of  0.26394858956336975 is Guessian.

```python
# I want to find out all variables with outliers except "Alcalinity of ash", because it is Guessian.
def find_outliers_IQR(df):
   q1 = df.quantile(0.25)
   q3 = df.quantile(0.75)
   IQR = q3 - q1
   outliers = df[((df < (q1 - 1.5 * IQR)) | (df > (q3 + 1.5 * IQR)))]
   return outliers

columns= ['Alcohol', 'Malic acid', 'Ash', 'Magnesium', 'Total phenols', 'Flavanoids', 
              'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines',
              'Proline']

outliers = find_outliers_IQR(df[columns])

print("number of outliers: " + str(len(outliers)))
print("max outlier value: " + str(outliers.max()))
print("min outlier value: " + str(outliers.min()))
```

```python
# Columns with outliers:
outlier_columns = outliers.columns[outliers.any()]
print("Columns with outliers:", ', '.join(outlier_columns))
```

```python
# Columns without outliers:
non_outlier_columns = outliers.columns[~outliers.any()]
print("Columns without outliers:", ', '.join(non_outlier_columns))
```

```python
from sklearn.metrics import accuracy_score

# Fit and evaluate each model
models = [model_1, model_2, model_3, model_4]
accuracies = []

for model in models:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    accuracies.append(accuracy)

# Print the accuracies for each model
for i, model in enumerate(models):
    print("Model", i+1, "Accuracy:", accuracies[i])
```

### confusion_matrix

```python
from sklearn.metrics import confusion_matrix

# Fit and evaluate each model
confusion_matrices = []

for model in models:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    cm = confusion_matrix(y_test, predictions)
    confusion_matrices.append(cm)

# Print the confusion matrices for each model
for i, model in enumerate(models):
    print("Model", i+1, "Confusion Matrix:")
    print(confusion_matrices[i])
    print()
```

- Model 3: Model 3 showed some inconsistencies in classifying the three classes. It misclassified two instances from class 2 as class 1 and misclassified one instance from class 1 as class 2. However, it performed well in classifying class 3.

- Model 4: Model 4 had a lower performance compared to the other models. It misclassified two instances from class 1 as class 2 and one instance from class 3 as class 1. However, it achieved high accuracy in classifying class 2.

Overall, Models 1 and 2 exhibited better performance compared to Models 3 and 4 in terms of accurate classification across the three classes. They had relatively fewer misclassifications and demonstrated more consistent performance.

Additionally, considering the specific requirements and objectives of your classification task will help in determining the most suitable model for your needs.

## Exercise 3
Train them using the different parameters they support in order to improve their prediction.

```python


from sklearn.model_selection import GridSearchCV


# Define the parameter grid for each model
param_grid_lr = {
    'C': [0.1, 1.0, 10.0],
    'penalty': ['l1', 'l2']
}

param_grid_svc = {
    'C': [0.1, 1.0, 10.0],
    'kernel': ['linear', 'rbf'],
    'gamma': [0.1, 1.0, 10.0]
}

param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree']
}

param_grid_dt = {
    'max_depth': [None, 5, 10],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [2, 5, 10]
}

# Create instances of the models
model_lr = LogisticRegression()
model_svc = SVC()
model_knn = KNeighborsClassifier()
model_dt = DecisionTreeClassifier()

# Perform GridSearchCV for each model
grid_search_lr = GridSearchCV(model_lr, param_grid=param_grid_lr, cv=5)
grid_search_svc = GridSearchCV(model_svc, param_grid=param_grid_svc, cv=5)
grid_search_knn = GridSearchCV(model_knn, param_grid=param_grid_knn, cv=5)
grid_search_dt = GridSearchCV(model_dt, param_grid=param_grid_dt, cv=5)

# Fit the models with GridSearchCV
grid_search_lr.fit(X_train, y_train)
grid_search_svc.fit(X_train, y_train)
grid_search_knn.fit(X_train, y_train)
grid_search_dt.fit(X_train, y_train)

# Get the best hyperparameters and the corresponding score for each model
best_params_lr = grid_search_lr.best_params_
best_score_lr = grid_search_lr.best_score_

best_params_svc = grid_search_svc.best_params_
best_score_svc = grid_search_svc.best_score_

best_params_knn = grid_search_knn.best_params_
best_score_knn = grid_search_knn.best_score_

best_params_dt = grid_search_dt.best_params_
best_score_dt = grid_search_dt.best_score_

# Print the best hyperparameters and score for each model
print("Logistic Regression - Best Hyperparameters:", best_params_lr)
print("Logistic Regression - Best Score:", best_score_lr)

print("SVC - Best Hyperparameters:", best_params_svc)
print("SVC - Best Score:", best_score_svc)

print("KNN - Best Hyperparameters:", best_params_knn)
print("KNN - Best Score:", best_score_knn)

print("Decision Tree - Best Hyperparameters:", best_params_dt)
print("Decision Tree - Best Score:", best_score_dt)
```
